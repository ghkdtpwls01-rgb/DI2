# 문제 1. 기초통계량 계산

**📘 문제**

- 온라인 쇼핑몰 고객 1,000명을 대상으로 1점부터 10점 사이의 만족도 조사를 실시했습니다.
  이 데이터를 바탕으로 고객 만족도의 **중심 경향성**과 **분포 특성**을 파악해 봅시다.

```python
# 데이터 생성
np.random.seed(2025)
satisfaction = np.random.normal(7.5, 1.5, 1000)
satisfaction = np.clip(satisfaction, 1, 10)
df1 = pd.DataFrame({'score': satisfaction})
```

**📌 아래를 수행해 보세요:**

- 평균, 중앙값, 최빈값, 표준편차를 계산해 봅시다.
- 히스토그램을 그리고 평균과 중앙값 선으로 표시해 봅시다.
- 데이터의 분포 형태를 해석한 후, 마케팅 전략 수립에 어떻게 활용할 수 있을지 생각해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 1] 여기에 코드를 작성해주세요.
# 중심 경향성 및 분포 특성 계산
mean_val = df1['score'].mean()
median_val = df1['score'].median()
mode_val = df1['score'].round(0).mode()[0]   # 반올림 후 최빈값
std_val = df1['score'].std()
skewness = df1['score'].skew()
kurtosis = df1['score'].kurt()

print("=== 고객 만족도 기술통계 ===")
print(f"평균(Mean): {mean_val:.3f}")
print(f"중앙값(Median): {median_val:.3f}")
print(f"최빈값(Mode, 반올림 기준): {mode_val:.1f}")
print(f"표준편차(Std): {std_val:.3f}")
print(f"왜도(Skewness): {skewness:.3f}")
print(f"첨도(Kurtosis): {kurtosis:.3f}")
print(f"최솟값: {df1['score'].min():.2f}, 최댓값: {df1['score'].max():.2f}")

# 히스토그램 시각화
plt.figure(figsize=(10, 6))
plt.hist(df1['score'], bins=20, color='#4C72B0', edgecolor='white', alpha=0.85)
plt.axvline(mean_val, color='#C44E52', linestyle='--', linewidth=2, label=f'평균: {mean_val:.2f}')
plt.axvline(median_val, color='#55A868', linestyle='-', linewidth=2, label=f'중앙값: {median_val:.2f}')
plt.title('온라인 쇼핑몰 고객 만족도 분포 (n=1,000)', fontsize=14, fontweight='bold')
plt.xlabel('만족도 점수 (1~10점)', fontsize=12)
plt.ylabel('고객 수', fontsize=12)
plt.legend(fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('satisfaction_histogram.png', dpi=150)
plt.show()

#2. 분포 형태 해석

#평균(7.43)과 중앙값(7.42)이 거의 일치 → 대칭에 가까운 종형(정규) 분포입니다. 원래 정규분포에서 생성했지만 1~10점으로 클리핑(clip)하면서 상한선 근처가 살짝 눌려 왜도가 약한 음수(-0.165)를 보입니다.
#첨도가 -0.274로 0에 가까움 → 정규분포보다 살짝 완만하지만 큰 차이는 없어, 극단적으로 뾰족하거나 납작하지 않은 무난한 분포입니다.
#표준편차 1.39로, 대다수 고객(약 68%)이 6.0~8.8점 사이에 분포 — 만족도가 전반적으로 양호하지만 편차도 존재합니다.

#3. 마케팅 전략에의 활용 방안

#평균 7.4점은 "양호하지만 최상은 아닌" 수준 → 8점 이상 우수 고객군에게는 리뷰 요청·리퍼럴(추천) 프로그램을 적용해 입소문 자산으로 전환.
#6점 이하 저만족 고객군(하위 약 16~20%): 이탈 위험군으로 분류해 개별 설문이나 쿠폰·사과 메일 등 리텐션 캠페인 타겟팅.
#분포가 대칭적이므로 평균값 하나만으로 전체 고객을 대표해도 큰 왜곡이 없어, 만족도 KPI 관리 시 평균 추적이 유효한 지표로 사용 가능.
#표준편차가 크지 않다는 것은 서비스 품질이 비교적 일관되다는 신호 — 반대로 표준편차가 컸다면 서비스 편차(특정 지점/시간대 문제)를 의심해봐야 했을 것입니다.
```

# 문제 2. 평균 vs 중앙값 비교

**📘 문제**

- A지역과 B지역의 연봉 데이터를 비교해 봅시다.
- B 지역에는 일부 고소득자가 포함되어 있습니다.

```python
# 데이터 생성
np.random.seed(2025)

# A 지역: 평균 300, 표준편차 50
region_a = np.random.normal(300, 50, 100)

# B 지역: 일반 소득자 + 극단적인 고소득자 포함
region_b = np.concatenate([
    np.random.normal(280, 40, 95),    # 일반 소득자 95명
    np.random.normal(1500, 200, 5)    # 고소득자 5명
])

df2 = pd.DataFrame({
    'region': ['A'] * 100 + ['B'] * 100,
    'salary': np.concatenate([region_a, region_b])
})
```

**📌 아래를 수행해 보세요:**

- 각 지역의 평균 연봉과 중앙값을 구해 봅시다.
- 지역별 박스플롯(Boxplot)을 그려서 분포의 차이를 시각적으로 확인해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 2] 여기에 코드를 작성해주세요.
# 지역별 평균 / 중앙값 계산
summary = df2.groupby('region')['salary'].agg(
    평균='mean', 중앙값='median', 표준편차='std',
    최솟값='min', 최댓값='max'
).round(2)
print("=== 지역별 연봉 기술통계 ===")
print(summary)

# 박스플롯 시각화
from matplotlib.lines import Line2D

fig, ax = plt.subplots(figsize=(8, 6))
data_to_plot = [df2[df2['region']=='A']['salary'], df2[df2['region']=='B']['salary']]
bp = ax.boxplot(data_to_plot, tick_labels=['A 지역', 'B 지역'], patch_artist=True,
                 widths=0.5, showmeans=True,
                 meanprops={'marker':'D', 'markerfacecolor':'red', 'markeredgecolor':'red', 'markersize':8})

colors = ['#4C72B0', '#DD8452']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_title('지역별 연봉 분포 비교 (Boxplot)', fontsize=14, fontweight='bold')
ax.set_ylabel('연봉 (만원)', fontsize=12)
ax.grid(axis='y', alpha=0.3)
ax.legend(handles=[
    Line2D([0], [0], marker='D', color='w', markerfacecolor='red', markersize=8, label='평균'),
    Line2D([0], [0], color='orange', lw=2, label='중앙값')
], loc='upper left', fontsize=10)

plt.tight_layout()
```

# 문제 3. 이상값과 분산 비교

# 문제 4. IQR 이상값 제거

**📘 문제**

- 앞의 문제에서 2반에는 **극단적인 이상값(예: 매우 낮은 점수)**이 포함되어 있어
  통계 지표, 특히 평균에 큰 영향을 주는 것을 확인했습니다.
- 이번에는 **IQR(사분위 범위)**을 기준으로 이상값을 제거한 뒤, 이상값 제거 전후의 평균을 비교해 봅시다

<br>

**📌 아래를 수행해 보세요:**

- 2반의 점수 데이터를 대상으로 IQR 기준을 사용해 이상값을 판별하고 제거해 봅시다.
- 이상값 제거 전과 제거 후의 평균을 각각 계산해 봅시다.
- 이상값 제거가 평균에 어떤 영향을 주는지 설명해 봅시다.
- 필요하다면 중앙값과도 비교해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 4] 여기에 코드를 작성해주세요.
class2_scores = df3[df3['class']=='2반']['score']

# IQR 기준 이상값 판별
q1 = class2_scores.quantile(0.25)
q3 = class2_scores.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(f'Q1: {q1:.2f}, Q3: {q3:.2f}, IQR: {iqr:.2f}')
print(f'이상값 판별 하한: {lower_bound:.2f}, 상한: {upper_bound:.2f}')

outliers = class2_scores[(class2_scores < lower_bound) | (class2_scores > upper_bound)]
print(f'탐지된 이상값: {sorted(outliers.tolist())}')

# 이상값 제거
cleaned = class2_scores[(class2_scores >= lower_bound) & (class2_scores <= upper_bound)]

print(f'제거 전 - 평균: {class2_scores.mean():.2f}, 중앙값: {class2_scores.median():.2f}, 표준편차: {class2_scores.std():.2f}')
print(f'제거 후 - 평균: {cleaned.mean():.2f}, 중앙값: {cleaned.median():.2f}, 표준편차: {cleaned.std():.2f}')
```

```python
# 데이터를 어떻게 읽을까요?
# [문제 4] 여기에 의견을 작성해주세요.
▸ 평균이 중앙값에 가까워졌는가? 아니요, 오히려 차이가 살짝 더 벌어졌습니다 (0.73 → 1.35). 얼핏 "이상값을 제거하면 평균이 중앙값에 가까워질 것"이라고 예상하기 쉽지만, 이번 사례는 그렇지 않았습니다.

▸ 어느 방향으로 영향을 주었는가? 평균은 73.73 → 74.35로 +0.62점 소폭 상승했습니다. 왜냐하면 제거된 두 이상값(0점, 130점)의 평균이 (0+130)/2 = 65로, 원래 2반 평균(73.73)보다 낮았기 때문입니다. 즉 이 두 값의 "합"이 전체 평균을 살짝 아래로 누르고 있었던 것이고, 제거하니 근소하게 위로 튀어 오른 겁니다. (0점과 130점이 완벽히 대칭이었다면 평균은 전혀 안 변했을 것입니다.)
▸ IQR이 데이터를 나누는 방식: 데이터를 크기 순으로 정렬한 뒤 4등분합니다.

Q1(1사분위수): 하위 25% 지점
Q3(3사분위수): 하위 75% 지점
IQR = Q3 − Q1 (중간 50% 데이터가 차지하는 폭)

▸ 이상값 정의 방식: Q1 - 1.5×IQR 보다 작거나 Q3 + 1.5×IQR 보다 큰 값을 이상값으로 판단합니다. (1.5는 통계학자 존 튜키가 제안한 관례적 배수)

우리 2반 데이터: Q1=60.75, Q3=88.37, IQR=27.61 → 정상 범위 19.33~129.79 → 이 범위를 벗어난 0점과 130점이 정확히 이상값으로 잡혔습니다.
▸ 평균/표준편차 기준과 비교했을 때 IQR이 더 강건한 이유

평균과 표준편차 기준으로 이상값을 판단하는 대표적 방법은 "평균 ± 2 표준편차" 또는 "평균 ± 3 표준편차"입니다. 문제는 평균과 표준편차 자체가 이미 이상값의 영향을 받는다는 점입니다.
   이번 경우엔 우연히 둘 다 잡혔지만, 표준편차 자체가 이상값 때문에 이미 21.19로 부풀려져 있다는 게 문제입니다. 만약 이상값이 하나 더 있었거나 더 극단적이었다면, 표준편차가 계속 커지면서 "이상값 판별 상한선"도 같이 넓어져 정작 이상값을 못 잡는 현상(이를 masking effect, 가려짐 효과라고 함)이 벌어질 수 있습니다.
   반면 IQR은 중앙 50% 데이터(Q1~Q3)만으로 계산되기 때문에, 바깥쪽에 이상값이 몇 개 있든 Q1과 Q3 자체는 거의 흔들리지 않습니다. 그래서 이상값이 있어도 "정상 범위"의 기준선이 왜곡되지 않는다는 점에서 **IQR이 평균/표준편차 기준보다 더 강건(robust)**합니다.
    제거해야 할 때 (오류·오염으로 판단되는 경우):
유지해야 할 때 (의미 있는 정보인 경우):

상황	더 유용한 지표	이유
데이터가 대칭적이고 이상값이 적을 때	평균	모든 값의 정보를 반영해 더 정밀한 대표값
이상값이나 극단적 쏠림이 있을 때	중앙값	극단값에 영향받지 않아 "전형적인 값"을 더 안정적으로 반영
소득, 자산, 집값처럼 오른쪽 꼬리가 긴 데이터	중앙값	"보통 사람"의 상황을 왜곡 없이 보여줌
평균 계산 자체가 필요한 경우(예: 총합 예측)	평균	중앙값으로는 총합을 복원할 수 없음 (평균×개수=총합이지만 중앙값은 불가)
▸ 이상값이 있을 때 어떤 지표가 더 신뢰할 수 있나: 중앙값입니다. 방금 확인했듯, 2반에서 이상값을 제거해도 중앙값은 73.00으로 전혀 변하지 않았지만, 평균은 흔들렸습니다. 이는 중앙값이 "튀는 값 몇 개"에 흔들리지 않는 성질을 보여줍니다.

▸ 두 지표를 함께 보는 것의 장점: 평균과 중앙값의 차이 자체가 하나의 진단 도구가 됩니다.

차이가 거의 없다 → 분포가 대칭적, 이상값 영향이 적음 (또는 서로 상쇄됨)
차이가 크다 → 분포가 한쪽으로 치우쳤거나 이상값이 존재할 가능성 → 추가 조사 필요
```

   - 범주형 (categorical)
   - 서열형 범주 (ordinal-categorical)

```python
# 아래를 수행해 보세요:
# [문제 5] 여기에 코드를 작성해주세요.
# 데이터 타입 확인
print(df5.dtypes)


# 서열형으로 알려진 범주와 순서 (사람이 미리 지정해줘야 하는 정보)
ordinal_orders = {
    'size': ['Small', 'Medium', 'Large'],
    'satisfaction': [1, 2, 3, 4, 5]
}

def classify_variable(series, col_name, ordinal_map, unique_ratio_threshold=0.05):
    dtype = series.dtype
    n_unique = series.nunique()
    n_total = len(series)
    unique_ratio = n_unique / n_total

    # 1. 문자열(object 또는 string)이면서 순서 정보가 없으면 -> 범주형(명목형)
    is_string_like = (dtype == object) or pd.api.types.is_string_dtype(series)
    if is_string_like and col_name not in ordinal_map:
        return '범주형 (categorical)'

    # 2. 순서 정보가 사전에 등록되어 있으면 -> 서열형 범주
    if col_name in ordinal_map:
        return '서열형 범주 (ordinal-categorical)'

    # 3. 숫자형인데 고유값이 매우 적으면(=거의 반복됨) -> 이산형으로 의심
    if pd.api.types.is_numeric_dtype(dtype):
        if unique_ratio < unique_ratio_threshold or n_unique <= 10:
            return '이산형 (numeric-discrete) - 서열형 여부 확인 필요'
        else:
            return '연속형 (numeric-continuous)'

    return '분류 불가 - 수동 확인 필요'

print(f"{'변수명':<15}{'dtype':<10}{'고유값개수':<10}{'고유비율':<10}{'유형 분류'}")
print("-" * 80)
for col in df5.columns:
    result = classify_variable(df5[col], col, ordinal_orders)
    n_unique = df5[col].nunique()
    ratio = n_unique / len(df5)
    print(f"{col:<15}{str(df5[col].dtype):<10}{n_unique:<10}{ratio:<10.3f}{result}")
```

2. 변수 유형에 따라 적절한 분석 방법이나 시각화 방식은 어떻게 달라질까요?

# 문제 9. 정규분포와 표준편차

**📘 문제**

- 정규분포와 표준편차의 개념을 실제 데이터에 적용해 보겠습니다.
- 정규분포를 따르는 점수 데이터에서 다음을 확인해 봅시다:
  - 평균 ±1σ, 평균 ±2σ 범위에 속한 값들의 비율은 실제로 얼마나 되는가?
  - 이 결과가 이론적인 68-95-99 법칙과 얼마나 유사한지 비교해 봅시다.

<br>

**📌 아래를 수행해 보세요:**

1. 평균이 70, 표준편차가 10인 정규분포 데이터를 1000개 생성해 봅시다.
2. 평균을 중심으로 ±1σ, ±2σ 범위에 포함되는 데이터의 비율을 계산해 봅시다.
3. 결과를 바탕으로 68-95-99 법칙과 비교해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 9] 여기에 코드를 작성해주세요.
np.random.seed(42)  # 재현성을 위해 시드 고정 (원하는 값으로 바꿔도 됨)
data = np.random.normal(70, 10, 1000)
mean = data.mean()
std = data.std()

print(f'실제 평균: {mean:.3f}, 실제 표준편차: {std:.3f}')

for k in [1, 2, 3]:
    lower = mean - k*std
    upper = mean + k*std
    within = ((data >= lower) & (data <= upper)).sum()
    ratio = within / len(data) * 100
    print(f'평균 ± {k}σ 범위 내 데이터 비율: {ratio:.2f}%')

# 시각화
plt.figure(figsize=(10,6))
plt.hist(data, bins=40, color='#4C72B0', edgecolor='white', alpha=0.8, density=True)

colors = ['#55A868', '#DD8452', '#C44E52']
for k, c in zip([1,2,3], colors):
    plt.axvline(mean - k*std, color=c, linestyle='--', alpha=0.7)
    plt.axvline(mean + k*std, color=c, linestyle='--', alpha=0.7, label=f'±{k}σ')
plt.axvline(mean, color='black', linestyle='-', linewidth=1.5, label='평균')

plt.title('정규분포 데이터와 68-95-99.7 법칙', fontsize=14, fontweight='bold')
plt.xlabel('값')
plt.ylabel('밀도')
plt.legend()
plt.tight_layout()
```

**🧠 데이터를 어떻게 읽을까요?**

1. ±1σ, ±2σ 범위에 실제로 몇 %의 데이터가 포함되었나요?
2. 이 결과는 68–95–99 법칙과 얼마나 비슷했나요?
3. 표준편차는 왜 중요한 지표인가요?
4. 이 개념이 실제로 어떻게 활용될 수 있을까요?

```python
# 데이터를 어떻게 읽을까요?
# [문제 9] 여기에 의견을 작성해주세요.
1. ±1σ, ±2σ 범위 내 실제 데이터 비율
±1σ (60.41 ~ 79.98): 68.60%
±2σ (50.62 ~ 89.77): 95.60%
매우 근접했습니다. 두 경우 모두 오차가 0.5%p 이내로, 사실상 이론값과 일치한다고 볼 수 있는 수준입니다. 이렇게 정확히 맞아떨어진 이유는 데이터를 애초에 np.random.normal()로 정규분포를 따르도록 생성했고, 표본 크기(1000개)도 충분히 커서 우연에 의한 오차가 크지 않았기 때문입니다.
    표준편차는 데이터가 평균을 중심으로 얼마나 퍼져 있는지를 하나의 숫자로 요약해줍니다. 단순히 "평균이 70이다"라는 정보만으로는 데이터가 60~80 사이에 몰려있는지, 아니면 0~140까지 넓게 퍼져있는지 전혀 알 수 없습니다. 표준편차가 있어야:

데이터의 신뢰성/일관성을 판단할 수 있고 (표준편차가 작으면 예측 가능성이 높음)
정규분포를 가정할 때, 특정 값이 얼마나 "흔한지" 또는 "드문지"를 정량적으로 계산할 수 있습니다 (68-95-99.7 법칙이 바로 이 계산의 근거)
서로 다른 두 데이터셋을 비교할 때, 평균이 같아도 표준편차가 다르면 완전히 다른 성격의 데이터라는 걸 알 수 있습니다 (예: 이전에 다룬 1반 vs 2반 성적 사례)
4. 품질관리, 이상거래 탐지, 성적 표준화, 신뢰구간 계산, A/B 테스트 등에서 "정상 범위"를 정하는 기준으로 활용됩니다.
이상치·품질관리: 공장에서 부품 규격이 평균 ±2σ를 벗어나면 불량으로 판정하는 식으로, 제조업의 품질관리(6시그마 등)에서 핵심적으로 쓰입니다.
이상 거래 탐지: 신용카드 사용 금액이 평소 패턴(평균 ±σ)에서 크게 벗어나면 이상 거래로 자동 플래그 처리하는 데 사용됩니다.
성적/시험 표준화: 원점수 대신 "평균에서 몇 표준편차 떨어져 있는가"를 나타내는 표준점수(Z-score)나 편차치를 계산해, 서로 다른 시험 간 성적을 비교할 수 있게 해줍니다.
신뢰구간 설정: 여론조사나 실험 결과에서 "95% 신뢰구간"을 계산할 때, 바로 이 ±2σ(정확히는 ±1.96σ) 개념이 사용됩니다.
A/B 테스트: 새로운 마케팅 전략의 효과가 우연한 변동 범위(예: ±2σ)를 벗어나는지 확인해, 그 차이가 통계적으로 유의미한지 판단하는 데 활용됩니다.
```

# 문제 10. 표본 평균 실험

**📘 문제**

- 큰 모집단에서 표본을 여러 번 추출하여 표본 평균의 분포를 비교해봅시다.

<br>

**📌 아래를 수행해 보세요:**

1. 평균이 100, 표준편차가 15인 모집단을 생성합니다.
2. 다음 두 조건에서 각각 1000번씩 표본 평균을 구합니다:
   - 표본 크기 5
   - 표본 크기 50
3. 두 조건의 **표본 평균 분포(histogram)**를 비교합니다.

```python
# 아래를 수행해 보세요:
# [문제 10] 여기에 코드를 작성해주세요.
np.random.seed(42)

# 1. 모집단 생성
population = np.random.normal(100, 15, 100000)

# 2. 표본 크기별로 1000번씩 표본 평균 계산
sample_means_5 = [np.random.choice(population, 5, replace=False).mean() for _ in range(1000)]
sample_means_50 = [np.random.choice(population, 50, replace=False).mean() for _ in range(1000)]

sample_means_5 = np.array(sample_means_5)
sample_means_50 = np.array(sample_means_50)

print(f'표본 크기5 - 표본평균의 평균: {sample_means_5.mean():.3f}, 표준오차: {sample_means_5.std():.3f}')
print(f'표본 크기50 - 표본평균의 평균: {sample_means_50.mean():.3f}, 표준오차: {sample_means_50.std():.3f}')

# 3. 히스토그램 비교
fig, axes = plt.subplots(1, 2, figsize=(14,6), sharex=True)

axes[0].hist(sample_means_5, bins=30, color='#4C72B0', edgecolor='white', alpha=0.85)
axes[0].axvline(100, color='red', linestyle='--', linewidth=2, label='모평균(100)')
axes[0].set_title(f'표본 크기 n=5 (표준편차={sample_means_5.std():.2f})', fontweight='bold')
axes[0].legend()

axes[1].hist(sample_means_50, bins=30, color='#DD8452', edgecolor='white', alpha=0.85)
axes[1].axvline(100, color='red', linestyle='--', linewidth=2, label='모평균(100)')
axes[1].set_title(f'표본 크기 n=50 (표준편차={sample_means_50.std():.2f})', fontweight='bold')
axes[1].legend()

fig.suptitle('표본 크기에 따른 표본평균 분포 비교 (중심극한정리)', fontsize=14, fontweight='bold')
plt.tight_layout()
```

# 문제 11. 단순 선형 회귀 분석

**📘 문제**

- 공부 시간과 성적 간의 관계를 회귀 분석으로 확인해봅시다.

```python
# 데이터 생성
from sklearn.linear_model import LinearRegression

np.random.seed(505)
study_time = np.random.normal(5, 2, 100)
score = 50 + study_time * 8 + np.random.normal(0, 5, 100)
df11 = pd.DataFrame({'study_time': study_time, 'score': score})
```

**📌 아래를 수행해 보세요:**

1. 공부 시간(`study_time`)과 성적(`score`)간의 관계를 단순 선형 회귀모델로 분석합니다.
2. 회귀 모델을 학습하고, 기울기(계수)와 절편을 출력해 봅시다.
3. 산점도 위에 회귀선을 시각화해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 11] 여기에 코드를 작성해주세요.
# 회귀모델 학습
X = df11[['study_time']]
y = df11['score']
model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(X, y)

print(f'기울기(계수): {slope:.3f}')
print(f'절편: {intercept:.3f}')
print(f'결정계수(R²): {r_squared:.3f}')

# 시각화
plt.figure(figsize=(9,6))
plt.scatter(df11['study_time'], df11['score'], alpha=0.6, color='#4C72B0', label='실제 데이터')

x_line = np.linspace(df11['study_time'].min(), df11['study_time'].max(), 100).reshape(-1,1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='#C44E52', linewidth=2.5, label=f'회귀선: y={slope:.2f}x+{intercept:.2f}')

plt.title('공부 시간과 성적의 관계 (단순 선형회귀)', fontsize=14, fontweight='bold')
plt.xlabel('공부 시간 (시간)')
plt.ylabel('성적 (점)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
```

# 문제 12. A/B 테스트 해석

**📘 문제**

- 두 그룹의 전환율 차이를 계산하고 해석해봅시다.

```python
# 데이터 생성
np.random.seed(606)
group_a = np.random.binomial(1, 0.12, 1000)
group_b = np.random.binomial(1, 0.15, 1000)
```

**📌 아래를 수행해 보세요:**

1. A 그룹과 B 그룹의 전환 데이터를 살펴봅니다.
2. 각 그룹의 전환율을 계산하고, 차이를 확인합니다.
3. 이 차이가 우연인지, 의미 있는 차이인지 생각해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 12] 여기에 코드를 작성해주세요.
from scipy import stats
from scipy.stats import chi2_contingency

np.random.seed(606)
group_a = np.random.binomial(1, 0.12, 1000)
group_b = np.random.binomial(1, 0.15, 1000)

n_a, n_b = len(group_a), len(group_b)
conv_a, conv_b = group_a.sum(), group_b.sum()
rate_a, rate_b = conv_a/n_a, conv_b/n_b

print(f'A그룹 전환율: {rate_a*100:.2f}% ({conv_a}/{n_a})')
print(f'B그룹 전환율: {rate_b*100:.2f}% ({conv_b}/{n_b})')

# 2표본 비율 z검정
pooled_p = (conv_a + conv_b) / (n_a + n_b)
se = np.sqrt(pooled_p*(1-pooled_p)*(1/n_a + 1/n_b))
z = (rate_b - rate_a) / se
p_value = 2 * (1 - stats.norm.cdf(abs(z)))
print(f'z통계량: {z:.3f}, p-value: {p_value:.4f}')

# 카이제곱 검정 (교차검증)
table = np.array([[conv_a, n_a-conv_a],[conv_b, n_b-conv_b]])
chi2, p_chi, dof, expected = chi2_contingency(table, correction=False)
print(f'카이제곱: {chi2:.3f}, p-value: {p_chi:.4f}')
```

1. 두 그룹의 전환율 차이는 얼마나 되나요?

3. 이 실험 결과로 어떤 결정을 내릴 수 있을까요?
4. A/B 테스트는 현실에서 어떻게 쓰일까요?

```python
# 데이터를 어떻게 읽을까요?
# [문제 12] 여기에 의견을 작성해주세요.
. 두 그룹의 전환율 차이
A그룹: 13.10% (131/1,000)
B그룹: 13.30% (133/1,000)
차이: 0.20%p — 매우 작은 차이입니다.
2. "차이가 의미 있다"는 것의 뜻

단순히 숫자가 다르다고 해서 "의미 있는 차이"는 아닙니다. 통계적으로 유의미하다는 건, 이 차이가 우연(무작위 변동)만으로는 설명하기 어려울 정도로 크다는 뜻입니다. p-value가 그 기준이 되는데, 우리 결과에서는 p-value = 0.895로 매우 커서, "두 그룹이 사실 같은 전환율을 가져도 이 정도 차이는 흔하게 우연히 발생할 수 있다"는 의미입니다. 즉 지금 나온 0.20%p 차이는 의미 있는 차이가 아니라 우연한 변동으로 봐야 합니다.

3. 이 실험 결과로 내릴 수 있는 결정
"B가 A보다 낫다"고 단정할 수 없습니다. 현재 데이터만으로는 두 전략(A/B) 사이에 통계적으로 유의미한 성과 차이가 없다고 봐야 합니다.
만약 B안이 A안보다 도입 비용이나 운영 부담이 더 크다면, 굳이 B로 전환할 근거가 부족합니다.
반대로 B안에 새로운 기능이나 장기적 이점(예: 브랜드 이미지 개선)이 있고 비용 부담이 적다면, 유의미한 차이는 없어도 다른 이유로 B를 선택할 수는 있습니다. (통계적 유의성만이 유일한 결정 기준은 아님)
가장 합리적인 다음 행동: 표본 크기를 늘려 실험을 더 오래 지속하거나 더 많은 사용자에게 노출시켜, 실제로 차이가 있는지(혹시 검정력 부족으로 못 잡아낸 건 아닌지) 재확인하는 것입니다.
4. A/B 테스트가 현실에서 쓰이는 방식
웹/앱 UI 개선: 버튼 색상, 문구, 레이아웃을 두 버전으로 나눠 클릭률·구매전환율 비교
마케팅 문구·이메일 캠페인: 제목이나 할인 문구를 다르게 보내 오픈율·클릭률 비교
가격 정책 테스트: 동일 상품을 두 가격대로 노출해 매출/전환율에 미치는 영향 확인
알고리즘·추천 시스템 개선: 새로운 추천 알고리즘이 기존보다 실제로 더 나은 성과(클릭, 체류시간)를 내는지 검증
제품 기능 출시 전 검증: 신규 기능을 일부 사용자에게만 노출해, 전체 출시 전에 실제 효과가 있는지 소규모로 먼저 확인

핵심 원칙: A/B 테스트는 "감"이나 "의견"이 아니라 실제 데이터로 결정을 내리게 해주지만, 오늘 다룬 것처럼 표본 크기가 충분하지 않으면 실제로 존재하는 차이도 놓칠 수 있다는 한계를 항상 염두에 둬야 합니다.
```
